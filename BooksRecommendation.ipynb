```python
import numpy as np
```

```python
arr2=df['  num_pages']
```

```python
list1=[]
for i in range(0, len(arr2)):
  try:
    list1.append(int(arr2[i]))
  except:
    df.drop(i,inplace=True)
    print(i)
    continue
```

```python
np_arr=np.array(list1)
```

```python
np_arr
```

```python
avg=np.mean(np_arr)
```

```python
df['  num_pages']=df['  num_pages'].astype(int)
df['average_rating']=df['average_rating'].astype(float)
```

```python
count1=df[df['  num_pages']>avg]['average_rating']
```

```python
arr=np.array(count1)
avg_rating_books_with_high_page=np.mean(arr)
avg_rating_books_with_high_page
```

```python
count2=df[df['  num_pages']<=avg]['average_rating']
```

```python
arr=np.array(count2)
avg_rating_books_with_low_page=np.mean(arr)
avg_rating_books_with_low_page
```

```python
if(avg_rating_books_with_high_page<avg_rating_books_with_low_page):
  print('books with fewer pages rated higher than those with large page counts')
else:
  print('books with higher pages rated higher than those with fewer page counts')
```
